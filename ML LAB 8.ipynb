```python
"""
McCulloch-Pitts (MP) Neuron Implementation
A simple artificial neuron model that can implement logic gates
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import product

class MPNeuron:
    """
    McCulloch-Pitts Neuron Model
    - Binary inputs and outputs (0 or 1)
    - Fixed weights (no learning)
    - Threshold-based activation
    """

    def __init__(self, num_inputs):
        """
        Initialize MP Neuron

        Parameters:
        num_inputs: Number of input features
        """
        self.num_inputs = num_inputs
        self.weights = None
        self.threshold = None

    def set_parameters(self, weights, threshold):
        """
        Set weights and threshold manually

        Parameters:
        weights: List of weights for each input
        threshold: Activation threshold
        """
        if len(weights) != self.num_inputs:
            raise ValueError(f"Expected {self.num_inputs} weights, got {len(weights)}")

        self.weights = np.array(weights)
        self.threshold = threshold

    def predict(self, X):
        """
        Predict output for given inputs

        Parameters:
        X: Input array (single input or batch)

        Returns:
        Binary output (0 or 1)
        """
        if self.weights is None or self.threshold is None:
            raise ValueError("Weights and threshold not set. Use set_parameters() first.")

        X = np.array(X)

        # Handle single input
        if X.ndim == 1:
            weighted_sum = np.dot(X, self.weights)
            return 1 if weighted_sum >= self.threshold else 0

        # Handle batch inputs
        weighted_sum = np.dot(X, self.weights)
        return (weighted_sum >= self.threshold).astype(int)

    def __repr__(self):
        return f"MPNeuron(inputs={self.num_inputs}, weights={self.weights}, threshold={self.threshold})"


# ==================== LOGIC GATE IMPLEMENTATIONS ====================

def implement_AND_gate():
    """Implement AND gate using MP Neuron"""
    print("=" * 50)
    print("AND GATE IMPLEMENTATION")
    print("=" * 50)

    # Create MP Neuron with 2 inputs
    neuron = MPNeuron(num_inputs=2)

    # Set weights and threshold for AND gate
    # Both inputs must be 1 for output to be 1
    neuron.set_parameters(weights=[1, 1], threshold=2)

    # Generate truth table
    inputs = list(product([0, 1], repeat=2))

    print(f"\nConfiguration: {neuron}")
    print("\nTruth Table:")
    print("-" * 30)
    print("X1\tX2\t|\tOutput")
    print("-" * 30)

    for x1, x2 in inputs:
        output = neuron.predict([x1, x2])
        print(f"{x1}\t{x2}\t|\t{output}")

    return neuron


def implement_OR_gate():
    """Implement OR gate using MP Neuron"""
    print("\n" + "=" * 50)
    print("OR GATE IMPLEMENTATION")
    print("=" * 50)

    neuron = MPNeuron(num_inputs=2)

    # At least one input must be 1 for output to be 1
    neuron.set_parameters(weights=[1, 1], threshold=1)

    inputs = list(product([0, 1], repeat=2))

    print(f"\nConfiguration: {neuron}")
    print("\nTruth Table:")
    print("-" * 30)
    print("X1\tX2\t|\tOutput")
    print("-" * 30)

    for x1, x2 in inputs:
        output = neuron.predict([x1, x2])
        print(f"{x1}\t{x2}\t|\t{output}")

    return neuron


def implement_NOT_gate():
    """Implement NOT gate using MP Neuron"""
    print("\n" + "=" * 50)
    print("NOT GATE IMPLEMENTATION")
    print("=" * 50)

    neuron = MPNeuron(num_inputs=1)

    # Negative weight inverts the input
    neuron.set_parameters(weights=[-1], threshold=0)

    inputs = [[0], [1]]

    print(f"\nConfiguration: {neuron}")
    print("\nTruth Table:")
    print("-" * 30)
    print("X1\t|\tOutput")
    print("-" * 30)

    for x in inputs:
        output = neuron.predict(x)
        print(f"{x[0]}\t|\t{output}")

    return neuron


def implement_NAND_gate():
    """Implement NAND gate using MP Neuron"""
    print("\n" + "=" * 50)
    print("NAND GATE IMPLEMENTATION")
    print("=" * 50)

    neuron = MPNeuron(num_inputs=2)

    # Negative weights with appropriate threshold
    neuron.set_parameters(weights=[-1, -1], threshold=-1)

    inputs = list(product([0, 1], repeat=2))

    print(f"\nConfiguration: {neuron}")
    print("\nTruth Table:")
    print("-" * 30)
    print("X1\tX2\t|\tOutput")
    print("-" * 30)

    for x1, x2 in inputs:
        output = neuron.predict([x1, x2])
        print(f"{x1}\t{x2}\t|\t{output}")

    return neuron


def implement_NOR_gate():
    """Implement NOR gate using MP Neuron"""
    print("\n" + "=" * 50)
    print("NOR GATE IMPLEMENTATION")
    print("=" * 50)

    neuron = MPNeuron(num_inputs=2)

    # Negative weights with lower threshold
    neuron.set_parameters(weights=[-1, -1], threshold=-2)

    inputs = list(product([0, 1], repeat=2))

    print(f"\nConfiguration: {neuron}")
    print("\nTruth Table:")
    print("-" * 30)
    print("X1\tX2\t|\tOutput")
    print("-" * 30)

    for x1, x2 in inputs:
        output = neuron.predict([x1, x2])
        print(f"{x1}\t{x2}\t|\t{output}")

    return neuron


def visualize_logic_gates():
    """Visualize all logic gates"""
    gates = {
        'AND': ([1, 1], 2),
        'OR': ([1, 1], 1),
        'NAND': ([-1, -1], -1),
        'NOR': ([-1, -1], -2)
    }

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.ravel()

    for idx, (gate_name, (weights, threshold)) in enumerate(gates.items()):
        neuron = MPNeuron(num_inputs=2)
        neuron.set_parameters(weights=weights, threshold=threshold)

        # Create grid for visualization
        x1 = np.array([0, 1, 0, 1])
        x2 = np.array([0, 0, 1, 1])

        outputs = []
        for i in range(4):
            outputs.append(neuron.predict([x1[i], x2[i]]))

        # Plot
        ax = axes[idx]
        colors = ['red' if o == 0 else 'green' for o in outputs]

        ax.scatter(x1, x2, c=colors, s=500, alpha=0.6, edgecolors='black', linewidth=2)

        for i in range(4):
            ax.annotate(f'{outputs[i]}', (x1[i], x2[i]),
                       fontsize=20, ha='center', va='center', fontweight='bold')

        ax.set_xlim(-0.5, 1.5)
        ax.set_ylim(-0.5, 1.5)
        ax.set_xlabel('X1', fontsize=12, fontweight='bold')
        ax.set_ylabel('X2', fontsize=12, fontweight='bold')
        ax.set_title(f'{gate_name} Gate\nWeights: {weights}, Threshold: {threshold}',
                    fontsize=14, fontweight='bold')
        ax.grid(True, alpha=0.3)
        ax.set_xticks([0, 1])
        ax.set_yticks([0, 1])

    plt.tight_layout()
    plt.savefig('mp_neuron_logic_gates.png', dpi=150, bbox_inches='tight')
    plt.show()

    print("\nVisualization saved as 'mp_neuron_logic_gates.png'")


def demonstrate_batch_prediction():
    """Demonstrate batch prediction"""
    print("\n" + "=" * 50)
    print("BATCH PREDICTION DEMO")
    print("=" * 50)

    neuron = MPNeuron(num_inputs=2)
    neuron.set_parameters(weights=[1, 1], threshold=2)  # AND gate

    # Multiple inputs at once
    X_batch = np.array([
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ])

    predictions = neuron.predict(X_batch)

    print(f"\nConfiguration: {neuron}")
    print("\nBatch Predictions:")
    print("-" * 30)

    df = pd.DataFrame(X_batch, columns=['X1', 'X2'])
    df['Output'] = predictions
    print(df.to_string(index=False))


# ==================== MAIN EXECUTION ====================

if __name__ == "__main__":
    print("\n" + "=" * 50)
    print("MP NEURON: LOGIC GATE IMPLEMENTATION")
    print("=" * 50)

    # Implement all logic gates
    and_gate = implement_AND_gate()
    or_gate = implement_OR_gate()
    not_gate = implement_NOT_gate()
    nand_gate = implement_NAND_gate()
    nor_gate = implement_NOR_gate()

    # Demonstrate batch prediction
    demonstrate_batch_prediction()

    # Visualize
    print("\n" + "=" * 50)
    print("GENERATING VISUALIZATIONS")
    print("=" * 50)
    visualize_logic_gates()

    print("\n" + "=" * 50)
    print("DEMO COMPLETE!")
    print("=" * 50)
```
