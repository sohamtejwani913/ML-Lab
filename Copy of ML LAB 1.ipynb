```python
import pandas as pd
import numpy as np

# Load CSV file
df = pd.read_csv('Bitcoin_history_data.csv')  # Replace with your actual file path

# 1. Mean of Closing Prices
mean_close_pd = df['Close'].mean()
mean_close_np = np.mean(df['Close'].values)

# 2. Maximum Volume Day
max_volume_day_pd = df[df['Volume'] == df['Volume'].max()]
max_index_np = np.argmax(df['Volume'].values)
max_volume_day_np = df.iloc[max_index_np]

# 3. Daily Price Range (High - Low)
df['Price_Range'] = df['High'] - df['Low']
price_range_np = np.subtract(df['High'].values, df['Low'].values)

# 4. Count of Days with Close > 400
count_above_400_pd = (df['Close'] > 400).sum()
count_above_400_np = np.sum(df['Close'].values > 400)

# 5. Top 3 Days by Opening Price
sorted_open_pd = df.sort_values(by='Open', ascending=False)
sorted_indices_np = np.argsort(-df['Open'].values)
sorted_open_np = df.iloc[sorted_indices_np]

# 6. Rolling 3-Day Mean of Close
df['Rolling_Close_3'] = df['Close'].rolling(window=3).mean()
rolling_close_np = np.convolve(df['Close'].values, np.ones(3)/3, mode='valid')

# 7. Standard Deviation of Volume
std_volume_pd = df['Volume'].std()
std_volume_np = np.std(df['Volume'].values)

# 8. Day with Minimum Low Price
min_low_day_pd = df[df['Low'] == df['Low'].min()]
min_index_np = np.argmin(df['Low'].values)
min_low_day_np = df.iloc[min_index_np]

# 9. Median of Close Prices
median_close_pd = df['Close'].median()
median_close_np = np.median(df['Close'].values)

# 10. Percentage Change in Closing Price
df['Pct_Change_Close'] = df['Close'].pct_change() * 100
pct_change_np = (df['Close'].values[1:] - df['Close'].values[:-1]) / df['Close'].values[:-1] * 100

# -------------------- PRINT RESULTS --------------------

print("1. Mean Close Price:")
print(f"   Pandas: {mean_close_pd}")
print(f"   NumPy : {mean_close_np}\n")

print("2. Max Volume Day:")
print("   Pandas:")
print(max_volume_day_pd)
print("   NumPy:")
print(max_volume_day_np, "\n")

print("3. Price Range (first 5 entries):")
print("   NumPy:", price_range_np[:5], "\n")

print("4. Days with Close > 400:")
print(f"   Pandas: {count_above_400_pd}")
print(f"   NumPy : {count_above_400_np}\n")

print("5. Top 3 Days by Opening Price:")
print("   Pandas:")
print(sorted_open_pd[['Date', 'Open']].head(3))
print("   NumPy:")
print(sorted_open_np[['Date', 'Open']].head(3), "\n")

print("6. Rolling 3-Day Mean of Close:")
print("   NumPy (first 5):", rolling_close_np[:5])
print("   Pandas (first 5):", df['Rolling_Close_3'].head(5).tolist(), "\n")

print("7. Standard Deviation of Volume:")
print(f"   Pandas: {std_volume_pd}")
print(f"   NumPy : {std_volume_np}\n")

print("8. Day with Minimum Low Price:")
print("   Pandas:")
print(min_low_day_pd)
print("   NumPy:")
print(min_low_day_np, "\n")

print("9. Median of Close Prices:")
print(f"   Pandas: {median_close_pd}")
print(f"   NumPy : {median_close_np}\n")

print("10. Percentage Change in Closing Price (first 5):")
print("   Pandas:", df['Pct_Change_Close'].head(5).tolist())
print("   NumPy :", pct_change_np[:5])
```
